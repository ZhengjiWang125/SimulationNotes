## Traditional Box Muller

```python
import numpy as np
import scipy.stats as spst
import scipy.special as spsp
import matplotlib.pyplot as plt
```

### Normal Distribution

```python
def sampling():
    U1 = spst.uniform.rvs()
    U2 = spst.uniform.rvs()
    R2 = -2*np.log(U1)
    R = np.sqrt(R2)
    theta = 2*np.pi*U2
    x1 = R * np.cos(theta)
    x2 = R * np.sin(theta)
    return (x1,x2)
```

```python
n = 10000
x  = np.zeros([n,2])
for i in range(0,n):
    x[i,:] = sampling()
    
samples = x.flatten()
```

```python
plt.hist(samples,density = True, bins = 100)
x = np.linspace(-4,4,1000)
plt.plot(x,spst.norm.pdf(x))
plt.show()
```

## Polar Method

```python
def sampling():
    U1 = spst.uniform.rvs()
    U2 = spst.uniform.rvs()
    V1 = 2*U1 - 1
    V2 = 2*U2 - 1
    S = V1 ** 2 + V2 ** 2
    while S > 1:
        U1 = spst.uniform.rvs()
        U2 = spst.uniform.rvs()
        V1 = 2*U1 - 1
        V2 = 2*U2 - 1
        S = V1 ** 2 + V2 ** 2
    c = np.sqrt(-2*np.log(S)/S)
    X = c* V1
    Y = c* V2
    return (X,Y)
```

```python
n = 10000
x = np.zeros([n,2])
for i in range(0,n):
    x[i,:] = sampling()
samples = x.flatten()
```

```python
x = np.linspace(-3,3)
plt.plot(x,spst.norm.pdf(x))
plt.hist(samples,density = True, bins = 100)
plt.show()
```

```python
import numpy as np
import simpy

from scipy.stats import expon, poisson, uniform
```

```python
def arrival(env,desk,doctor,arrivals,departures):
    while True:
        dt = expon.rvs(scale = 1/7)
        yield env.timeout(dt)
        arrivals.append(env.now)
        env.process(service(env,desk,doctor,arrivals,departures))

def service(env,desk,doctor,arrivals,departures):
    rqt = desk.request()
    yield rqt
    yield env.timeout(expon.rvs(scale = 1/7))
    desk.release(rqt)
    rqt = doctor.request()
    yield rqt
    yield env.timeout(expon.rvs(scale = 1/7))
    doctor.release(rqt)
    departures.append(env.now)



def system():
    env = simpy.Environment()
    desk = simpy.Resource(env,capacity = 1)
    doctor = simpy.Resource(env,capacity = 1)
    arrivals=  []
    departures = []
    env.process(arrival(env,desk,doctor,arrivals,departures))
    env.run(until = 1)
    return arrivals, departures
```

```python
def arrival(env,stations,cooks,revenues):
    i = 0
    while True:
        dt = expon.rvs(scale = 1/10)
        if env.now + dt > 8:
            break
        else:
            yield env.timeout(dt)
            print(f"customer {i} arrives")
            i += 1
            env.process(service(env,stations,cooks,revenues,i))
            
def service(env,stations,cooks,revenues,i):
    rqt = stations.request()
    yield rqt
    if env.now > 8:
        stations.release(rqt)
    else:
        yield env.timeout(expon.rvs(scale = 1/10))
        stations.release(rqt)
        rqt = cooks.request()
        yield rqt
        dt = expon.rvs(scale = 1/10)
        yield env.timeout(dt)
        cooks.release(rqt)
        revenues.append(10 * dt)


def system():
    env = simpy.Environment()
    stations = simpy.Resource(env,capacity = 3)
    cooks = simpy.Resource(env,capacity = 3)
    revenues = []
    env.process(arrival(env,stations,cooks,revenues))
    env.run()
    return np.sum(revenues)

def eval_revenue():
    np.random.seed(42)
    samples = [system() for i in range(100)]
    return np.mean(samples),np.std(samples)
```

## Multi-sever in parallel: one line, you don't pick the line..... Also not the same resources

```python
def arrival(env,desk1,desk2,doctor1,doctor2):
    i = 0
    while True:
        dt = expon.rvs(scale = 1/20)
        yield env.timeout(dt)
        env.process(service(env,desk1,desk2,doctor1,doctor2,i))
        print(f"customer {i} arrives at {env.now}.")
        i += 1

def service(env,desk1,desk2,doctor1,doctor2,i):
    rqt1 = desk1.request()
    rqt2 = desk2.request()
    result = yield rqt1 | rqt2
    if rqt1 in result and rqt2 not in result:
        print(f"customer {i} goes to desk1 at {env.now}")
        rqt2.cancel()
        dt = expon.rvs(scale = 1/6)
        yield env.timeout(dt)
        desk1.release(rqt1)
    elif rqt2 in result and rqt1 not in result:
        print(f"customer {i} goes to desk2 at {env.now}")
        rqt1.cancel()
        dt = expon.rvs(scale = 1/4)
        yield env.timeout(dt)
        desk2.release(rqt2)  
    else:
        print(f"customer {i} goes to desk1 at {env.now}")
        desk2.release(rqt2)
        dt = expon.rvs(scale = 1/6)
        yield env.timeout(dt)
        desk1.release(rqt1)
    rqt1 = doctor1.request()
    rqt2 = doctor2.request()
    result = yield rqt1 | rqt2
    if rqt1 in result and rqt2 not in result:
        print(f"customer {i} goes to doctor1 at {env.now}")
        rqt2.cancel()
        dt = expon.rvs(scale = 1/5)
        yield env.timeout(dt)
        doctor1.release(rqt1)
    elif rqt2 in result and rqt1 not in result:
        print(f"customer {i} goes to doctor2 at {env.now}")
        rqt1.cancel()
        dt = expon.rvs(scale = 1/3)
        yield env.timeout(dt)
        doctor2.release(rqt2)   
    else:
        print(f"customer {i} goes to doctor1 at {env.now}")
        doctor2.release(rqt2)
        dt = expon.rvs(scale = 1/5)
        yield env.timeout(dt)
        doctor1.release(rqt1)
    print(f"customer {i} finished at {env.now}")



def system():
    env = simpy.Environment()
    desk1 = simpy.Resource(env,capacity = 1)
    desk2 = simpy.Resource(env,capacity = 1)
    doctor1 = simpy.Resource(env, capacity = 1)
    doctor2 = simpy.Resource(env, capacity = 1)
    env.process(arrival(env,desk1,desk2,doctor1,doctor2))
    env.run(until = 1)
```

```python
'''
what customer would do is they would first look at how many people are lining up for the server
len(server.queue) + server.count
The request would only be created for the server with the shortest queue.
'''
```

customer arrives, spends some time picking the items, spend some time queuing, and spend some time checking out.

```python
def arrival(env,counters,customers):
    i = 0
    while True:
        dt = expon.rvs(scale = 1/25)
        yield env.timeout(dt)
        print(f"customer {i} has arrived at {env.now}")
        customers.append(1)
        env.process(service(env,counters,customers,i))
        i += 1
        

def service(env,counters,customers,i):
    yield env.timeout(expon.rvs(scale = 1/25))
    print (f"customer {i} just determined the items at {env.now}")
    lens = [len(counter.queue) + counter.count for counter in counters]
    ind = np.argmin(lens)
    rqt = counters[ind].request()
    yield rqt
    yield env.timeout(expon.rvs(scale = 1/20))
    counters[ind].release(rqt)
    print(f"customer {i} finished and went out at {env.now}")
    customers.append(-1)
    
    



def system():
    env = simpy.Environment()
    customers = []
    counters = [simpy.Resource(env,capacity = 1) for i in range(3)]
    env.process(arrival(env,counters,customers))
    env.run(until = 1)
    return customers
```

```python
'''
Suppose we have store.
(1) We have two kinds of customers: 
    (a) Regular Customers lambda_1
    (b) Return Customers making returns lambda_2
(2) N1 checkout lines, N2 customer service lines
    return only goes to customer service
    regular goes to whatever it wants
(3) regular lambda_3
    return lambda_4
(4) system will run until T, NO ARRIVAL! but we finish processing people in line
'''
```

```python
np.random.seed(42)
params = {'N_1': 4, 'N_2':1,'lambda_1':30, 'lambda_2': 10, 'lambda_3':35, 'lambda_4': 8}


def arrival(env,counter,checkouts,cuservices):
    i = 0
    while True:
        dt1 = expon.rvs(scale = 1/params['lambda_1'])
        dt2 = expon.rvs(scale = 1/params['lambda_2'])
        if dt1 < dt2: #this means regular customer comes
            if env.now + dt1 > 8:
                break
            yield env.timeout(dt1)
            print(f" a regular customer {i} comes at {env.now}")
            env.process(regular_service(env,counter,checkouts,cuservices, i))
        else:
            if env.now + dt2 > 8:
                break
            yield env.timeout(dt2)
            print(f" a returning customer {i} comes at {env.now}")
            env.process(return_service(env,counter,checkouts,cuservices, i))
        i += 1
        
def regular_service(env,counter,checkouts,cuservices, i):
    checkout_lens = [(len(i.queue) + i.count) for i in checkouts] 
    cuservices_lens = [(len(j.queue) + j.count) for j in cuservices]
    checkout_ind = np.argmin(checkout_lens)
    cuservices_ind = np.argmin(cuservices_lens)
    if min(checkout_lens) <= min(cuservices_lens): #this means the customer would go to checkout 
        rqt = checkouts[checkout_ind].request()
        yield rqt
        yield env.timeout(expon.rvs(scale = 1/params['lambda_3']))
        print(f" a regular customer {i} finished with a checkout at {env.now}")
        checkouts[checkout_ind].release(rqt)
    else:
        counter.append(i)
        rqt = cuservices[cuservices_ind].request()
        yield rqt
        yield env.timeout(expon.rvs(scale = 1/params['lambda_3']))
        print(f" a regular customer {i} finished with a customer service at {env.now}")
        cuservices[cuservices_ind].release(rqt)
    
    
    
def return_service(env,counter,checkouts,cuservices, i):
    cuservices_lens = [(len(i.queue) + i.count) for i in cuservices]
    cuservices_ind = np.argmin(cuservices_lens)
    rqt = cuservices[cuservices_ind].request()
    yield rqt
    yield env.timeout(expon.rvs(scale = 1/params['lambda_4']))
    print(f" a return customer {i} finished with a customer service at {env.now}")
    cuservices[cuservices_ind].release(rqt)






def system():
    env = simpy.Environment()
    counter = []
    checkouts = [simpy.Resource(env,capacity = 1) for i in range(params['N_1'])]
    cuservices = [simpy.Resource(env,capacity = 1) for i in range(params['N_2'])]
    env.process(arrival(env,counter,checkouts,cuservices))
    env.run()
    return counter



```

```python
'''
Suppose we have store.
(1) get rid of the customer service desks/customers
(2) each regular spens a random time lambda_2, checkout is 10 *Zi *Yi
(3) arriving customers will turn around and leave if more than M in store
(4) Customers in the store at close time are permited to finish
'''
```

```python

def arrival(env,leave,store,checkouts):
    i = 0
    while True:
        dt = expon.rvs(scale = 1/30)
        if env.now + dt >8:
            break
        yield env.timeout(dt)
        if store.count < 15:
            print(f"customer {i} has arrived")
            env.process(service(env,leave,store,checkouts,i))
        else:
            print(f"customer {i} leaves beacuse he has social anxiety")
            leave.append(i)
        i = i + 1    

def service(env,leave,store,checkouts,i):
    rqt = store.request()
    Z = expon.rvs(scale = 1/10)
    yield env.timeout(Z)
    lens = [checkout.count + len(checkout.queue) for checkout in checkouts]
    ind = np.argmin(lens)
    rqt1 = checkouts[ind].request()
    yield rqt1
    Y = expon.rvs(scale = 1/10)
    dt = 10 * Y * Z
    yield env.timeout(dt)
    checkouts[ind].release(rqt1)
    store.release(rqt)

def system():
    np.random.seed(42)
    env = simpy.Environment()
    leave = []
    store = simpy.Resource(env,capacity = 15)
    checkouts = [simpy.Resource(env,capacity = 1) for _ in range(4)]
    env.process(arrival(env,leave,store,checkouts))
    env.run()
    return leave
    
    

def evaluate_leave():
    samples = [system() for _ in range(100)]
    return np.mean(samples)


    
```

```python
def arrival(env,leave,store,checkouts,snackbar):
    i = 0
    while True:
        dt = expon.rvs(scale = 1/30)
        if env.now + dt >8:
            break
        yield env.timeout(dt)
        groupnum = 1 + poisson.rvs(3)
        if store.count + groupnum <= 15:
            print(f"group {i} has arrived")
            env.process(service(env,leave,store,checkouts,snackbar,i,groupnum))
        else:
            print(f"group {i} leaves beacuse he has social anxiety")
            leave.append(groupnum)
        i = i + 1    

def service(env,leave,store,checkouts,snackbar,i,groupnum):
    store_rqt = [store.request() for _ in range(groupnum)]
    Z = expon.rvs(scale = 1/10)
    yield env.timeout(Z)
    processes = [env.process(checkout_process(env,leave,store,checkouts,snackbar,i,groupnum,Z))]
    if groupnum > 1:
        processes.append(env.process(snackbar_process(env,leave,store,checkouts,snackbar,i,groupnum)))
    yield simpy.AllOf(env,processes)
    for rqt in store_rqt:
        store.release(rqt)

        
def checkout_process(env,leave,store,checkouts,snackbar,i,groupnum,Z):
    lens = [len(checkout.queue) + checkout.count for checkout in checkouts]
    ind = np.argmin(lens)
    rqt = checkouts[ind].request()
    yield rqt
    Y = expon.rvs(scale = 1/10)
    dt = 10 * Y * Z
    yield env.timeout(dt)
    checkouts[ind].release(rqt)
    
    
    
    
def snackbar_process(env,leave,store,checkouts,snackbar,i,groupnum):
    rqt = snackbar.request()
    yield rqt
    dt = expon.rvs(scale = 1/ 10)
    yield env.timeout(dt)
    snackbar.release(rqt)

        
        
        

def system():
    np.random.seed(42)
    env = simpy.Environment()
    leave = []
    store = simpy.Resource(env,capacity = 15)
    checkouts = [simpy.Resource(env,capacity = 1) for _ in range(4)]
    snackbar = simpy.Resource(env, capacity = 2)
    env.process(arrival(env,leave,store,checkouts,snackbar))
    env.run()
    return leave
    
    

def evaluate_leave():
    samples = [system() for _ in range(100)]
    return np.mean(samples)

```

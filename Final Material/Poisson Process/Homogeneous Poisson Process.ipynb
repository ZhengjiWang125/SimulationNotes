### There are two ways to generate homogeneous Poisson Process

### 1. use exponential

```python
'''
1. Let t = 0, i = 1
2. generate dt from exponential distribution
3. if t + dt > T, return N,S
4. else t = t + dt N = N + 1, Si = t, i = i + 1
'''
```

```python
import numpy as np
from scipy.stats import expon, poisson, uniform, binom
```

```python
def poisson_process(lmbda,T):
    N = 0
    t = 0
    S = []
    dt = expon.rvs(scale = 1/lmbda)
    while t + dt < T:
        t = t + dt
        N += 1
        S.append(t)
        dt = expon.rvs(scale = 1/lmbda)
    return N,np.array(S)
```

```python
def evaluate_process(S,t):
    return np.sum(S<=t)
```

```python
lmbda = 1.5
T = 5
nsims = 10000
t_vals = [0,1,2,3,4,5]
Nt = np.zeros((nsims, len(t_vals)))
for i in range(0,nsims):
    N,S = poisson_process(lmbda,T)
    for j,t in enumerate(t_vals):
        Nt[i,j] = evaluate_process(S,t)
Nt_diff = Nt[:,1:] - Nt[:,:-1]
print('Poisson process means: ',np.mean(Nt_diff, axis = 0))
print('Poisson process covariance: ', np.cov(Nt_diff, rowvar = False))
```

### 2. use ordered statistics

```python
'''
1.Generate poisson rv N
2.Generate N Uniform rvs, time them by T
3. sort and return
'''
```

```python
def poisson_process(lmbda,T):
    N = poisson.rvs(lmbda * T)
    S = uniform.rvs(size = N) * T
    return N, np.sort(S)
```

```python
def evaluate_process(S,t):
    return np.sum(S<=t)
```

```python
lmbda = 1.5
T = 5
nsims = 10000
t_vals = [1,4]
Nt = np.zeros((nsims,len(t_vals)))
Ns = np.zeros(nsims)
for i in range(0,nsims):
    N,S = poisson_process(lmbda,T)
    Ns[i] = N
    for j,t in enumerate(t_vals):
        Nt[i,j] = evaluate_process(S,t)
```

```python
import matplotlib.pyplot as plt
k_vals = [6,7,8,9]
fig, ax = plt.subplots(nrows = 4, ncols = 2, figsize = (10,10))
for i, k in enumerate(k_vals):
    for j, t in enumerate(t_vals):
        loc_k = (Ns == k)
        Nt_temp = Nt[loc_k,j]
        vals,counts = np.unique(Nt_temp, return_counts = True)
        ax[i,j].bar(vals,counts/len(Nt_temp))
        x_vals = np.arange(0,k+1)
        theo_vals = binom.pmf(x_vals, n=k, p= t/T)
        ax[i,j].plot(x_vals, theo_vals)
        
        
```
